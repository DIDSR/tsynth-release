The following notebook is used to visualize inference results and compare models for synthetic test set results. 
First we need to train detection models using ```synth``` and ```DM_synth``` for ```--experiment``` flag to run training for DBT and DM synthetic images. Inference results must first be collected on a set of models using ```evaluate_detectors.py```

Make sure to run: ```python download_scripts/download_results.py --all``` from ```code/``` directory before running this script.

````python
import os
import sys
sys.path.append(os.path.dirname(os.getcwd())) # add upper code/ directory to path 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utils_plot import plot_results, process_results
import config_global
````

Specify number of readers, the paths of the results csv files that were generated for each run by ```evaluate_detectors.py```, and the range of false positive values (per image) to plot. First we will check out DBT results.

````python
num_readers = 5
root_dir = config_global.dir_global + 'data/results/DBT/synth' 
result_paths = [f'{root_dir}/{i}/result.csv' for i in range(1, num_readers+1)]
avg_fpr = np.linspace(0, 5, 100)
````

Below DBT results for four different types of **breast densities**: fatty, scattered, heterogenious and dense.

````python
fig, ax = plt.subplots(figsize=(4.5, 3.5))

bd_values = [1, 2, 3, 4]
bd_labels = ['Fatty', 'Scattered', 'Heterogeneous', 'Dense']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#9467bd']

ax = plot_results(ax, result_paths, key='breast_density', key_values=bd_values, key_labels=bd_labels, colors=colors, avg_fpr=avg_fpr)

# Set axis labels with improved font settings
legend = ax.legend(title="Breast Density", fontsize=9, title_fontsize=10, loc='lower right', frameon=True)
legend.get_frame().set_edgecolor('black')
plt.tight_layout()
# Save figure for high-resolution submission
# plt.savefig('images/froc_bd_dbt.png', bbox_inches='tight', dpi=300)
plt.show()
````

Below DBT results separated by **lesion size** 5.0, 7.0 and 9.0

````python
fig, ax = plt.subplots(figsize=(4.5, 3.5))

ls_values = [9.0, 7.0, 5.0]
ls_labels = [ '9.0', '7.0', '5.0']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#9467bd']

ax = plot_results(ax, result_paths, key='lesion_size', key_values=ls_values, key_labels=ls_labels, colors=colors, avg_fpr=avg_fpr)

# Set axis labels with improved font settings
legend = ax.legend(title="Lesion Size (mm)", fontsize=9, title_fontsize=10, loc='lower right', frameon=True)
legend.get_frame().set_edgecolor('black')
plt.tight_layout()
# Save figure for high-resolution submission
# plt.savefig('images/froc_ls_dbt.png', bbox_inches='tight', dpi=300)
plt.show()
````
